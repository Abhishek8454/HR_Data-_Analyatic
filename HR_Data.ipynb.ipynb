**Q.6) How many employees Resigned & Terminated in each Department?**

```python
r_emp = df[df["Status"]=="Resigned"].groupby("Department")["Status"].count()
```

```python
t_emp = df[df["Status"]=="Terminated"].groupby("Department")["Status"].count()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

plt.bar(r_emp.index, r_emp.values, color="black", label="Resigned")
plt.bar(t_emp.index, t_emp.values, color="orange", label="Terminated", alpha=0.7)

plt.xlabel("Department")
plt.ylabel("Count of Employees")
plt.title("Employee Status by Department")
plt.legend()
plt.grid()
plt.show()
```
